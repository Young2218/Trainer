```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler

fv = vlo_10.fillna(-1)

mm = MinMaxScaler().fit_transform(fv.drop(columns=["Date"]).values)
sd = StandardScaler().fit_transform(fv.drop(columns=["Date"]).values)
rb = RobustScaler().fit_transform(fv.drop(columns=["Date"]).values)
ma = MaxAbsScaler().fit_transform(fv.drop(columns=["Date"]).values)

plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.title("StandardScaler")
plt.hist(sd.reshape(-1,1), bins=30)

plt.subplot(2,2,2)
plt.title("MinMaxSclaer")
plt.hist(mm.reshape(-1,1), bins=30)

plt.subplot(2,2,3)
plt.title("RobustScaler")
plt.hist(rb.reshape(-1,1), bins=30)

plt.subplot(2,2,4)
plt.title("MaxAbsScaler")
plt.hist(ma.reshape(-1,1), bins=30)
plt.show()
```
